# Skin screener runner

Notebook for `skin_homog/skin_screener.py`. Use `RUN_MODE` for `create` or `merge`, `FETCH_MODE` for runtime mix or a single sort, and `TARGET_UNIQUE` to override runtime JSON when needed. Logs stream directly into the cell output.


```python
from pathlib import Path

REPO = Path.cwd()
if not (REPO / 'skin_homog' / 'skin_screener.py').exists():
    REPO = Path('C:/Roman/skins_roundtrip_v1')

SKIN_LIST = REPO / 'lists' / 'skins_normal_filtered1.py'
OUTPUT_DIR = REPO / 'skin_homog' / 'data_skins_big'
RUN_MODE = 'merge'  # 'create' | 'merge'
FETCH_MODE = 'runtime'  # 'runtime' | 'mix' | single CSFloat sort
TARGET_UNIQUE = None  # None -> use runtime JSON
SKIP_KNOWN_IDS = None  # None -> use runtime JSON
IGNORE_EXISTING_ITEMS = None  # None -> use runtime JSON
RUNTIME_JSON = REPO / 'skin_homog' / 'skin_screener_runtime.json'
```

```python
import importlib.util
import os

assert RUN_MODE in {'create', 'merge'}

runner_path = REPO / 'skin_homog' / 'skin_screener.py'
spec = importlib.util.spec_from_file_location('skin_screener', runner_path)
skin_screener = importlib.util.module_from_spec(spec)
spec.loader.exec_module(skin_screener)

argv = [str(runner_path)]
argv.append('--create' if RUN_MODE == 'create' else '--merge')
if FETCH_MODE == 'mix':
    argv.append('--mix')
elif FETCH_MODE != 'runtime':
    argv.extend(['--no-mix', '--sort', str(FETCH_MODE)])
if TARGET_UNIQUE is not None:
    argv.extend(['--target', str(TARGET_UNIQUE)])
if SKIP_KNOWN_IDS is True:
    argv.append('--skip-known-ids')
elif SKIP_KNOWN_IDS is False:
    argv.append('--allow-known-ids')
if IGNORE_EXISTING_ITEMS is True:
    argv.append('--ignore-existing-items')
elif IGNORE_EXISTING_ITEMS is False:
    argv.append('--allow-existing-items')
argv.append(str(SKIN_LIST))

os.environ['SKIN_SCREENER_OUTPUT_DIR'] = str(OUTPUT_DIR)
os.environ['SKIN_SCREENER_RUNTIME_CONFIG'] = str(RUNTIME_JSON)

items = skin_screener.load_items(str(SKIN_LIST))
print(f'repo   {REPO}')
print(f'items  {len(items)} from {SKIN_LIST.name}')
print(f'mode   {RUN_MODE}')
print(f'fetch  {FETCH_MODE}')
print(f'skip_known_ids {SKIP_KNOWN_IDS}')
print(f'ignore_existing_items {IGNORE_EXISTING_ITEMS}')
print(f'out    {OUTPUT_DIR}')
print(f'cfg    {RUNTIME_JSON}')
print('argv   ' + ' '.join(argv))
```

```python
exit_code = skin_screener.run_cli(argv)
print(f'\nexit_code={exit_code}')
if exit_code != 0:
    raise SystemExit(exit_code)
```
