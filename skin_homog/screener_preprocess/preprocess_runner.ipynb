# Preprocess screener runner

Cheap first-pass scan for many skins. Collects `base_price`, capped `n_listings`, and mean `1 - ask/predicted` on a small listing sample.

```python
from pathlib import Path

REPO = Path.cwd()
if not (REPO / 'skin_homog' / 'screener_preprocess' / 'preprocess_screener.py').exists():
    REPO = Path('C:/Roman/skins_roundtrip_v1')

SKIN_LIST = REPO / 'lists' / 'skins_normal.py'
RUN_MODE = 'merge'  # 'create' | 'merge'
SORT_BY = None  # None -> use runtime JSON
LISTINGS_CAP = None  # None -> use runtime JSON
DISCOUNT_SAMPLE_LIMIT = None  # None -> use runtime JSON
OUTPUT_CSV = REPO / 'skin_homog' / 'screener_preprocess' / 'preprocess_metrics.csv'
RUNTIME_JSON = REPO / 'skin_homog' / 'screener_preprocess' / 'preprocess_runtime.json'
```

```python
import importlib.util
import os
import sys

runner_path = REPO / 'skin_homog' / 'screener_preprocess' / 'preprocess_screener.py'
spec = importlib.util.spec_from_file_location('preprocess_screener', runner_path)
preprocess_screener = importlib.util.module_from_spec(spec)
spec.loader.exec_module(preprocess_screener)

argv = [str(runner_path), '--create' if RUN_MODE == 'create' else '--merge']
if SORT_BY is not None:
    argv.extend(['--sort', str(SORT_BY)])
if LISTINGS_CAP is not None:
    argv.extend(['--cap', str(LISTINGS_CAP)])
if DISCOUNT_SAMPLE_LIMIT is not None:
    argv.extend(['--sample', str(DISCOUNT_SAMPLE_LIMIT)])
argv.extend(['--output', str(OUTPUT_CSV)])
argv.append(str(SKIN_LIST))

os.environ['PREPROCESS_SCREENER_RUNTIME_CONFIG'] = str(RUNTIME_JSON)

print(f'repo   {REPO}')
print(f'items  {SKIN_LIST}')
print(f'mode   {RUN_MODE}')
print(f'sort   {SORT_BY}')
print(f'cap    {LISTINGS_CAP}')
print(f'sample {DISCOUNT_SAMPLE_LIMIT}')
print(f'out    {OUTPUT_CSV}')
print(f'cfg    {RUNTIME_JSON}')
print('argv   ' + ' '.join(argv))
```

```python
old_argv = sys.argv[:]
try:
    sys.argv = argv
    rc = preprocess_screener.main()
finally:
    sys.argv = old_argv

print(f'rc={rc}')
```
