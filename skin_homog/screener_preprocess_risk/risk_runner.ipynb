# Risk preprocess runner

Stage-2 enrich for shortlist items from the cheap preprocess. Reads a shortlist `ITEMS` file, copies base rows from stage-1 preprocess CSV, then adds Steam trade/risk metrics.

```python
from pathlib import Path

REPO = Path.cwd()
if not (REPO / 'skin_homog' / 'screener_preprocess_risk' / 'risk_preprocess.py').exists():
    REPO = Path('C:/Roman/skins_roundtrip_v1')

SHORTLIST = REPO / 'lists' / 'skins_normal_filtered1.py'
STAGE1_CSV = REPO / 'skin_homog' / 'screener_preprocess' / 'preprocess_metrics.csv'
RUN_MODE = 'merge'  # 'create' | 'merge'
TRADE_DAYS = None  # None -> use runtime JSON
MIN_DISCOUNT_SAMPLE_N = None  # None -> use runtime JSON
OUTPUT_CSV = REPO / 'skin_homog' / 'screener_preprocess_risk' / 'risk_metrics.csv'
RUNTIME_JSON = REPO / 'skin_homog' / 'screener_preprocess_risk' / 'risk_runtime.json'
```

```python
import importlib.util
import os
import sys

runner_path = REPO / 'skin_homog' / 'screener_preprocess_risk' / 'risk_preprocess.py'
spec = importlib.util.spec_from_file_location('risk_preprocess', runner_path)
risk_preprocess = importlib.util.module_from_spec(spec)
spec.loader.exec_module(risk_preprocess)

argv = [str(runner_path), '--create' if RUN_MODE == 'create' else '--merge']
if TRADE_DAYS is not None:
    argv.extend(['--days', str(TRADE_DAYS)])
if MIN_DISCOUNT_SAMPLE_N is not None:
    argv.extend(['--min-discount-sample', str(MIN_DISCOUNT_SAMPLE_N)])
argv.extend(['--stage1-csv', str(STAGE1_CSV)])
argv.extend(['--output', str(OUTPUT_CSV)])
argv.append(str(SHORTLIST))

os.environ['RISK_PREPROCESS_RUNTIME_CONFIG'] = str(RUNTIME_JSON)

print(f'repo   {REPO}')
print(f'shortlist {SHORTLIST}')
print(f'stage1_csv {STAGE1_CSV}')
print(f'mode   {RUN_MODE}')
print(f'days   {TRADE_DAYS}')
print(f'min_discount_sample {MIN_DISCOUNT_SAMPLE_N}')
print(f'out    {OUTPUT_CSV}')
print(f'cfg    {RUNTIME_JSON}')
print('argv   ' + ' '.join(argv))
```

```python
old_argv = sys.argv[:]
try:
    sys.argv = argv
    rc = risk_preprocess.main()
finally:
    sys.argv = old_argv

print(f'rc={rc}')
```
