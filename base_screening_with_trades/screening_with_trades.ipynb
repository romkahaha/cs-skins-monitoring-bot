```python
# –°–Ї—А–Є–љ–Є–љ–≥ + —Б—В–∞—В–Є—Б—В–Є–Ї–∞ —Б–і–µ–ї–Њ–Ї SCM (/market/pricehistory) –Ј–∞ SCM_TRADE_DAYS –≤ fetchers.py
# –†–µ–Ї–Њ–Љ–µ–љ–і—Г–µ—В—Б—П: USE_EUR_PRICES=True вАФ steam_ask –Є SCM –≤ EUR (Steam), Float USD√ЧFrankfurter/ECB.
# –Ф–ї—П pricehistory: env STEAM_COOKIES –Є–ї–Є local_steam_cookies.py (STEAM_COOKIES)
from pathlib import Path
import sys

_here = Path.cwd().resolve()
for p in [_here, _here.parent]:
    if (p / "item_lists").is_dir():
        sys.path.insert(0, str(p))
        break

ITEM_LIST_FILE = "../lists/screening_full.py"
SAVE_DIR = "../data_with_trades"
USE_EUR_PRICES = True
WRITE_MODE = "create"  # create | merge
OUT_CSV = ""  # for merge set the existing csv path; for create blank => timestamped file
```

```python
from pathlib import Path
from datetime import datetime

from item_lists import load_item_list
from fetchers import fetch_all_prices_with_trades

ITEMS = load_item_list(ITEM_LIST_FILE)
print(f"{len(ITEMS)} items from {ITEM_LIST_FILE}\n")

name = Path(ITEM_LIST_FILE).stem
mode = str(WRITE_MODE).strip().lower()
if OUT_CSV:
    out = Path(OUT_CSV)
elif mode == "create":
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    out = Path(SAVE_DIR) / f"{name}_trades_{ts}.csv"
else:
    raise ValueError("Set OUT_CSV to an existing/future csv path when WRITE_MODE='merge'")

df = fetch_all_prices_with_trades(
    ITEMS,
    prices_in_eur=USE_EUR_PRICES,
    out_csv=out,
    write_mode=mode,
)

print("Saved", out.resolve())
df
```
